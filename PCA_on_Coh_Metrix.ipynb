# **Feature Preprocessing**

```python
#remove missing value
df.dropna(subset=['Reponse'],inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
df.head()
```

```python
# select CM features
# DESWC_3: total word written
# DESSLd_7: sentence length, number of words, SD
# DESWLsy_8: Word length number of syllables, mean
# PCVERBp_18: Text Easability PC deep cohesion
# LSASS1_38: LSA overlap, adjacent sentences, mean
# LDTTRc_46: Lexical diversity, TTR
# LDMTLD_48: Lexical diversity, MTLD
# CNCAll_50
# CNCCaus_51
# CNCLogic_52
# CNCADC_53
# CNCTempx_55
# CNCAdd_56
# CNCPos_57
# CNCNeg_58
# SMCAUSv_59
# SMCAUSvp_60
# SMINTEp_61
# SMCAUSr_62
# SMINTEr_63
# SMCAUSlsa_64
# SMCAUSwn_65
# SMTEMP_66
# SYNLE_67
# SYNSTRUTa_73
# WRDFRQc_92
# WRDCNCc_97
#====================================27 measures
# WRDNOUN
# WRDVERB
# WRDADJ
# WRDADV
# WRDPRO
# WRDPRP1s
# WRDPRP3s
# WRDFAMc
# WRDIMGc
# WRDMEAc
# CRFCWO1
# CRFCWO1d
# CRFAO1
# CRFAOa
# CRFNO1
# CRFSOa
#====================================43 measures

CM_df = df.loc[:, ["DESWC_3","DESSLd_7", "DESWLsy_8", "PCVERBp_18", "LSASS1_38", "LDTTRc_46",
                      "LDMTLD_48", "CNCAll_50", "CNCCaus_51","CNCLogic_52", "CNCADC_53", "CNCTempx_55",
                      "CNCAdd_56", "CNCPos_57", "CNCNeg_58", "SMCAUSv_59", "SMCAUSvp_60",
                      "SMINTEp_61", "SMCAUSr_62", "SMINTEr_63", "SMCAUSlsa_64", "SMCAUSwn_65",
                      "SMTEMP_66", "SYNLE_67", "SYNSTRUTa_73", "WRDFRQc_92", "WRDCNCc_97"]]
                      #"WRDNOUN", "WRDVERB", "WRDADJ", "WRDADV", "WRDPRO", "WRDPRP1s",
                      #"WRDPRP3s", 'WRDFAMc', "WRDIMGc", "WRDMEAc", "CRFCWO1", "CRFCWO1d",
                      #"CRFAO1", "CRFAOa", "CRFNO1", "CRFSOa"
CM_df
```

```python
#standardize/Normalize Data
num_cols = CM_df.columns[(CM_df.dtypes == 'float64') | (CM_df.dtypes == 'int64')]
num_cols
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(CM_df)
```

```python
scaled_data = scaler.transform(CM_df)
```

```python
scaled_data
```

# **Perform PCA on Coh-Metrix**

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pca = PCA(n_components=3)
pca.fit(scaled_data)
```

```python
CM_pca = pca.transform(scaled_data)
```
